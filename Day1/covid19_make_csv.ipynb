```python
import pandas as pd
import matplotlib as plt
import numpy as np

# 이 데이터는 누적 확진자 수를 보여줌
covid_con = pd.read_csv("time_series_covid19_confirmed_global.csv")
covid_con.set_index("Country/Region", inplace = True)
korea_covid = covid_con.loc['Korea','1/22/20':]

kcon_covid19 = pd.DataFrame(korea_covid)

# 누적 확진자 수 라고 칼럼 명을 바꿈
kcon_covid19 = kcon_covid19.rename(columns = {"Korea":"acc_decide"})
kcon_covid19

# 인덱스 바꿔줌 
index2 = []
for i in range(0,355) :
    index2.append(i)

kcon_covid19.index = index2[:]
kcon_covid19
```

```python
from datetime import date
from datetime import timedelta

# 날짜 2020-01-22 ~ 2021-01-10
start = date(2020, 1, 22)
end = date(2021, 1, 10)

print(end- start)

index = []
for i in range (0,355) :
    date = start + timedelta(i)
    index.append(date)

print(index[-1])
```

```python
# 0~ 354 인덱스 / 날짜 / 칼럼 ...
kcon_covid19 = kcon_covid19
kcon_covid19['date'] = index
kcon_covid19
```

```python
#초기화
for i in range(0,354) :
    kcon_covid19['decide_per_day'] = 0

kcon_covid19['decide_per_day'][0] = 0

# 일별 확진자 수 구하기
for i in range(1,355) :
    kcon_covid19['decide_per_day'][i] = kcon_covid19['acc_decide'][i] - kcon_covid19['acc_decide'][i-1]

kcon_covid19
```

```python
new = {'acc_decide' : [1, 1], 'date' : ['2020-01-20', '2020-01-21'], 'decide_per_day' : [1, 0]}
new = pd.DataFrame(new)
new
```

```python
con_covid19 = pd.concat([new,kcon_covid19])
con_covid19
```

```python
covid_dea = pd.read_csv("time_series_covid19_deaths_global.csv")
covid_dea.set_index("Country/Region", inplace = True)
covid_dea = covid_dea.loc['Korea, South','1/22/20':]

covid_dea = pd.DataFrame(covid_dea)

# 누적 확진자 수 라고 칼럼 명을 바꿈
kdea_covid19 = covid_dea.rename(columns = {"Korea, South":"acc_death"})

# 인덱스 바꿔줌 
kdea_covid19.index = index2[:]
kdea_covid19

# 0~ 354 인덱스 / 날짜 / 칼럼 ...
kdea_covid19 = kdea_covid19
kdea_covid19['date'] = index
kdea_covid19
```

```python
#초기화
for i in range(0,354) :
    kdea_covid19['death_per_day'] = 0

for i in range(1,355) :
    kdea_covid19['death_per_day'][i] = kdea_covid19['acc_death'][i] - kdea_covid19['acc_death'][i-1]

kdea_covid19
```

```python
new2 = {'acc_death' : [0, 0], 'date' : ['2020-01-20', '2020-01-21'], 'death_per_day' : [0, 0]}
new2 = pd.DataFrame(new2)
new2
```

```python
dea_covid19 = pd.concat([new2,kdea_covid19])
dea_covid19
```

```python
covid_re = pd.read_csv("time_series_covid19_recovered_global.csv")
covid_re.set_index("Country/Region", inplace = True)
covid_re = covid_re.loc['Korea, South','1/22/20':]

covid_re = pd.DataFrame(covid_re)

# 누적 확진자 수 라고 칼럼 명을 바꿈
kre_covid19 = covid_re.rename(columns = {"Korea, South":"acc_recovered"})

# 인덱스 바꿔줌 
kre_covid19.index = index2[:]
kre_covid19

# 0~ 354 인덱스 / 날짜 / 칼럼 ...
kre_covid19 = kre_covid19
kre_covid19['date'] = index
kre_covid19
```

```python
#초기화
for i in range(0,354) :
    kre_covid19['recovered_per_day'] = 0

for i in range(1,355) :
    kre_covid19['recovered_per_day'][i] = kre_covid19['acc_recovered'][i] - kre_covid19['acc_recovered'][i-1]

kre_covid19
```

```python
new3 = {'acc_recovered' : [0, 0], 'date' : ['2020-01-20', '2020-01-21'], 'recovered_per_day' : [0, 0]}
new3 = pd.DataFrame(new3)
new3
```

```python
re_covid19 = pd.concat([new3,kre_covid19])
re_covid19
```

```python
# 세 개의 데이터를 합침
covid19 = pd.merge(con_covid19, re_covid19)
covid19 = pd.merge(covid19, dea_covid19)

# 순서 정리
covid19 = covid19[['date','acc_decide','acc_recovered','acc_death','decide_per_day','recovered_per_day','death_per_day']]
covid19
```

```python
#csv파일로 추출
covid19.to_csv("covid19.csv", index=True)
```
